### Importación de datos



```python
import pandas as pd
import  matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

store1 = pd.read_csv(url)
store2 = pd.read_csv(url2)
store3 = pd.read_csv(url3)
store4 = pd.read_csv(url4)

store1
```

#1. Análisis de facturación y costo de envío



###Facturacion total

```python
store1FactTotal= store1['Precio'].sum()
store2FactTotal= store2['Precio'].sum()
store3FactTotal= store3['Precio'].sum()
store4FactTotal= store4['Precio'].sum()


print (f' Ingreso total de la tienda 1 : ' , store1['Precio'].sum())
print (f' Ingreso total de la tienda 2 : ' , store2['Precio'].sum())
print (f' Ingreso total de la tienda 3 : ' , store3['Precio'].sum())
print (f' Ingreso total de la tienda 4 : ' , store4['Precio'].sum())
```

###Grafico para representar las ventas totales por negocio


```python
#se crean 2 listas para las etiquetas de los ejes
tiendas=['Tienda 1', 'Tienda 2' , 'Tienda 3 ', 'Tienda 4 ']
ingresos=[store1FactTotal, store2FactTotal, store3FactTotal, store4FactTotal]

plt.barh(tiendas , ingresos , color=colores)
plt.title('Ventas totales por tienda')
plt.xlabel('Tiendas')
plt.ylabel('Ventas totales')
plt.show()
```

```python

store1['Fecha de Compra'] = pd.to_datetime(store1['Fecha de Compra'], dayfirst=True)
store2['Fecha de Compra'] = pd.to_datetime(store2['Fecha de Compra'], dayfirst=True)
store3['Fecha de Compra'] = pd.to_datetime(store3['Fecha de Compra'], dayfirst=True)
store4['Fecha de Compra'] = pd.to_datetime(store4['Fecha de Compra'], dayfirst=True)

# Agrupar las ventas por mes (puedes cambiar 'M' por 'W' para semana, 'D' para días, etc.)
ventas_store1 = store1.groupby(store1['Fecha de Compra'].dt.to_period('M')).agg({'Precio': 'sum'})
ventas_store2 = store2.groupby(store2['Fecha de Compra'].dt.to_period('M')).agg({'Precio': 'sum'})
ventas_store3 = store3.groupby(store3['Fecha de Compra'].dt.to_period('M')).agg({'Precio': 'sum'})
ventas_store4 = store4.groupby(store4['Fecha de Compra'].dt.to_period('M')).agg({'Precio': 'sum'})

# Graficar la evolución de ventas por tienda
plt.figure(figsize=(10, 6))
plt.plot(ventas_store1.index.astype(str), ventas_store1['Precio'], marker='o', label='Tienda 1', color=colores[0])
plt.plot(ventas_store2.index.astype(str), ventas_store2['Precio'], marker='o', label='Tienda 2', color=colores[1])
plt.plot(ventas_store3.index.astype(str), ventas_store3['Precio'], marker='o', label='Tienda 3', color=colores[2])
plt.plot(ventas_store4.index.astype(str), ventas_store4['Precio'], marker='o', label='Tienda 4', color=colores[3])

# Personalizar el gráfico
plt.title('Evolución de Ventas por Tienda a lo Largo del Tiempo')
plt.xlabel('Fecha (Mes)')
plt.ylabel('Ventas Totales')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

El presente informe tiene como objetivo asesorar al Sr. Juan en la toma de decisiones estratégicas respecto a sus tiendas de e-commerce. Dado su interés en iniciar un nuevo negocio, se busca determinar cuál de las cuatro tiendas actuales presenta menor desempeño general y, por tanto, sería la más recomendable para vender.

Este análisis se sustenta en métricas clave como: ingresos totales, categorías de productos más y menos vendidas, calificaciones promedio de clientes, productos destacados por volumen de ventas, y costo promedio de envío.

Después de evaluar el rendimiento financiero, la experiencia del cliente, la variedad de productos y los costos operativos, se concluye que la **Tienda 4 es la opción más adecuada para ser vendida**.
